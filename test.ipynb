```python
twitter_data_file = '/home/nuclear/Stuff/PythonPrograms/Twitter-Portfolio-Project/data/roe-wade-10000-original-sentiment-added.json'
twitter_data = pd.read_json(twitter_data_file)
twit_data_obj = tda.tweet_analysis(twitter_data)
```

```python
twit_data_obj.df = twit_data_obj.df.sort_values(by='tweet_time')
count_positive=0
pos_lst = []
count_negative=0
neg_lst = []
for row,sent in enumerate(twit_data_obj.df['net_score']):
    if sent.lower() == 'positive':
        count_positive += 1
        pos_lst.append(count_positive)
    elif sent.lower() != 'positive':
        pos_lst.append(count_positive)
    if sent.lower() == 'negative':
        count_negative += 1
        neg_lst.append(count_negative)
    elif sent.lower() != 'negative':
        neg_lst.append(count_negative)
print('Positive list:')
print(len(pos_lst))
# print(pos_lst)
print('Negative list:')
print(len(neg_lst))
# print(neg_lst)

twit_data_obj.df['neg_tweet_count'] = neg_lst
twit_data_obj.df['pos_tweet_count'] = pos_lst

# print(neg_lst.sort(reverse=True))
# print(neg_lst[:10])

twit_data_obj.df.head(2)
```
